```python
import numpy as np
from anml.prior.prior import Prior, UniformPrior
from anml.prior.prior_getter import SplinePriorGetter
from xspline import XSpline
```

```python
p = SplinePriorGetter(UniformPrior(lb=1.0, ub=2.0), domain_type="abs")
spline = XSpline(knots=np.linspace(0.0, 2.0, 5), degree=2)
mat1 = p.get_prior(spline).mat
```

```python
p = SplinePriorGetter(UniformPrior(lb=1.0, ub=2.0), domain_type="rel")
spline = XSpline(knots=np.linspace(0.0, 2.0, 5), degree=2)
mat2 = p.get_prior(spline).mat
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(mat1)
plt.plot(mat2)
```
